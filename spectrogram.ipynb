```python
from scipy.fft import fft, ifft, fftshift, fftfreq
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,2,0.001)
f0 = 50
f1 = 250
t1 = 2
x = np.cos(2 * np.pi * t *(f0 + (f1 - f0) * t**2/(3*t1**2)))
x_fnk = lambda t: np.cos(2 * np.pi * t *(f0 + (f1 - f0) * t**2/(3*t1**2)))
```

```python
fig,axs = plt.subplots(1 ,figsize=(20,2))
axs = plt.plot(t,x_fnk(t))
plt.ylabel('Amplitude')
plt.xlabel('Time [sec]')
plt.title('Time Domain')
```

```python
f, t, Sxx = signal.spectrogram(x,1e3)
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.title('Spectorgram')
plt.show()
```

```python
N = x.shape[0]
yf = fft(x)
xf = fftfreq(N)[:N//2]
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')
plt.title('Frequency Space')
plt.show()
```
